```python
state_gdp
total_gdp=state_gdp[(state_gdp['Items  Description']=='GSDP - CURRENT PRICES (` in Crore)')]
nunique = total_gdp.apply(pd.Series.nunique)
cols_to_drop = nunique[nunique == 1].index
total_gdp=total_gdp.drop(cols_to_drop, axis=1)
total_gdp=total_gdp.T

new_header = total_gdp.iloc[0] 
total_gdp = total_gdp[1:] 
total_gdp.columns = new_header 
total_gdp
```

```python
total_gdp = pd.DataFrame(total_gdp)
union_territories=['Andaman & Nicobar Islands','Chandigarh','Delhi','Puducherry','All_India GDP']
total_gdp=total_gdp.drop(union_territories) 
```

```python

total_gdp= total_gdp.reset_index()
total_gdp.head()
```

```python
#for year 2015 and 2016
gdp_1516= total_gdp[['index','2015-16']]
gdp_1516= gdp_1516.fillna(0)

gdp_1516= gdp_1516.sort_values(by=['2015-16'], ascending=False)
gdp_1516
```

```python
fig = plt.figure(figsize=(30, 15))
plt.bar(gdp_1516['index'], gdp_1516['2015-16'])
plt.xlabel('States', fontsize=15)
plt.ylabel('growth rate', fontsize=15)
plt.xticks(gdp_1516['index'], fontsize=15, rotation=90)
plt.title('Growth of each state in 2015-16')
plt.savefig("Growth of each state in 2015-16.png")
plt.show()
```
